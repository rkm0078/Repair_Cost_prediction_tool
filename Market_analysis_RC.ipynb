```python
import numpy as np
import matplotlib.pyplot as plt

# Given data
years = np.array([2020, 2021, 2022, 2023, 2024, 2025])
market_size = np.array([30000, 32250, 34662, 37249, 40023, 42999])  # Crore INR

# Financial model parameters
m = 50000  # Price per unit service in INR
c = 100 * 1e7  # Total fixed costs in Crore INR

# Market size in INR
market_size_inr = market_size * 1e7  # Convert crore to INR

# Calculate profit for each year
profit = m * market_size_inr - c

# Plotting the market size and profit forecast
plt.figure(figsize=(12, 6))

# Market size plot
plt.subplot(1, 2, 1)
plt.plot(years, market_size, 'o-', label='Market Size (Crore INR)')
plt.xlabel('Year')
plt.ylabel('Market Size (Crore INR)')
plt.title('Market Size Forecast')
plt.grid(True)
plt.legend()

# Profit plot
plt.subplot(1, 2, 2)
plt.plot(years, profit / 1e7, 's-', label='Profit (Crore INR)')
plt.xlabel('Year')
plt.ylabel('Profit (Crore INR)')
plt.title('Profit Forecast')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

profit
```
